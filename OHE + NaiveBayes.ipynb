```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder

from sklearn.naive_bayes import BernoulliNB, GaussianNB
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['sex', 'smoker', 'day']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder =  LabelEncoder()
lbl_encoder
```

```python
lbl_encoder.fit_transform(df[['time']])
```

```python
df['time_label'] = lbl_encoder.fit_transform(df[['time']])
```

```python
df.head()
```

```python
new_df = df.drop(columns=['sex','smoker','day', 'time'])
new_df.head()
```

```python
X = new_df.iloc[:,:-1]
y = new_df.iloc[:,-1]
```

# Model Scaling

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Model Training

```python
from sklearn.model_selection import GridSearchCV

parameters = {
'alpha' : [0.1,1.0,1.10,1.5,10]
}


classifier = BernoulliNB()


clf = GridSearchCV(classifier, param_grid=parameters, verbose=3, scoring='accuracy', cv=5 )
clf
```

```python
clf.fit(X_train_scaled, y_train)
```

```python
clf.best_params_
```

```python
model = BernoulliNB(alpha=0.1)

model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))

print('--------------------------------------------------------')

print(accuracy_score(y_test, y_pred))

print('-------------------------------------------------------')
print(classification_report(y_test, y_pred))
```
