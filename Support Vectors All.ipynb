# Support Vector Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
svc
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print(confusion_matrix(y_test, y_pred))

print('--------------------------------------------------------')

print(accuracy_score(y_test, y_pred))

print('-------------------------------------------------------')
print(classification_report(y_test, y_pred))
```

# HyperParameter with SVC

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'C' : [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel' :  ['linear']
}
```

```python
svc = SVC()
```

```python
grid = GridSearchCV(svc, param_grid=parameters, refit=True, cv=5, verbose=3)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))

print('--------------------------------------------------------')

print(accuracy_score(y_test, y_pred))

print('-------------------------------------------------------')
print(classification_report(y_test, y_pred))
```

# Support Vector Regressor

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000, n_features=2, n_targets=1, noise=3.0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear')
svr
```

```python
svr.fit(X_train, y_train)
```

```python
svr.coef_
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)
score
```

# HyperParameter with SVR

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'C' : [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel' :  ['linear'],
    'epsilon' : [0.1, 0.2, 0.3]          
}
```

```python
svr = SVR()
```

```python
grid = GridSearchCV(svr, param_grid=parameters, scoring = 'neg_mean_squared_error', refit=True, verbose=3, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = r2_score(y_test, y_pred)
score
```

# Support Vectors Kernals 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2 - x**2)

y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)

y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1, x1)
```

```python
import pandas as pd

df1 = pd.DataFrame(np.vstack([y,x]).T, columns=['X1', 'X2'])
df1['Y'] = 0

df2 = pd.DataFrame(np.vstack([y1,x1]).T, columns=['X1', 'X2'])
df2['Y'] = 1


df = df1.append(df2)
df.head()
```

```python
df.tail()
```

```python
df['X1_square'] = df['X1']**2
df['X2_square'] = df['X2']**2
df['X1_X2'] = df['X1']*df['X2']
```

```python
df.head()
```

```python
X = df[['X1_square', 'X2_square', 'X1_X2' ]]
y = df['Y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```
